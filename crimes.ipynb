```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
from shapely.geometry import Point
```

```python
crimes = pd.read_csv('data/locations/crimes.csv')
crimes['Geometry'] = crimes.apply(lambda r: Point(r[['Longitude', 'Latitude']]), axis=1)
crimes = gpd.GeoDataFrame(crimes).set_geometry('Geometry').set_crs(epsg=4326).to_crs(epsg=26986)
crimes
```

```python
crimes['OFFENSE_CODE_GROUP'].value_counts()
```

```python
crimes['OCCURRED_ON_DATE'] = pd.to_datetime(crimes['OCCURRED_ON_DATE'])
NUM_YEARS = (crimes['OCCURRED_ON_DATE'].max() - crimes['OCCURRED_ON_DATE'].min()).days / 365.25
NUM_YEARS
```

```python
vehicle_accidents = crimes.loc[crimes['OFFENSE_CODE_GROUP'] == 'Motor Vehicle Accident Response']
other_crimes = crimes.loc[crimes['OFFENSE_CODE_GROUP'] != 'Motor Vehicle Accident Response']
```

```python
open_spaces = pd.read_csv('data/locations/open_spaces.csv')
open_spaces['Geometry'] = open_spaces['Geometry'].apply(wkt.loads)
open_spaces = gpd.GeoDataFrame(open_spaces).set_geometry('Geometry').set_crs(epsg=4326).to_crs(epsg=26986)
open_spaces['Buffer'] = open_spaces.buffer(500)
open_spaces = open_spaces.set_geometry('Buffer')
open_spaces['Buffer Area'] = open_spaces['Buffer'].area / 4046.86
open_spaces
```

```python
open_spaces['Crimes'] = open_spaces['Geometry'].apply(lambda x: (other_crimes['Geometry'].distance(x) <= 500).sum())
open_spaces['Vehicle Accidents'] = open_spaces['Geometry'].apply(lambda x: (vehicle_accidents['Geometry'].distance(x) <= 500).sum())
open_spaces
```

```python
open_spaces['Nearby Crimes Per Acre Per Year'] = (open_spaces['Crimes'] / NUM_YEARS) / open_spaces['Buffer Area']
open_spaces['Nearby Vehicle Accidents Per Acre Per Year'] = (open_spaces['Vehicle Accidents'] / NUM_YEARS) / (open_spaces['Buffer Area'] - open_spaces['Acres'])
```

```python
open_spaces.to_csv('data/locations/open_spaces_with_crimes.csv', index=False)
```
